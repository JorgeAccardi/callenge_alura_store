### Importación de datos



```python
import pandas as pd

# URLs de los archivos CSV
url_tienda1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_tienda2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_tienda3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_tienda4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los DataFrames
tienda1_df = pd.read_csv(url_tienda1)
tienda2_df = pd.read_csv(url_tienda2)
tienda3_df = pd.read_csv(url_tienda3)
tienda4_df = pd.read_csv(url_tienda4)

# Mostrar las primeras filas de cada DataFrame para verificar la carga
print("Primeras filas de tienda1:")
print(tienda1_df.head())
```

#1. Análisis de facturación


```python
# Calcular el ingreso total para cada tienda
ingreso_total_tienda1 = tienda1_df['Precio'].sum()
ingreso_total_tienda2 = tienda2_df['Precio'].sum()
ingreso_total_tienda3 = tienda3_df['Precio'].sum()
ingreso_total_tienda4 = tienda4_df['Precio'].sum()

```

```python
# Mostrar los resultados
print(f"Ingreso total de la Tienda 1: ${ingreso_total_tienda1:,.2f}")
print(f"Ingreso total de la Tienda 2: ${ingreso_total_tienda2:,.2f}")
print(f"Ingreso total de la Tienda 3: ${ingreso_total_tienda3:,.2f}")
print(f"Ingreso total de la Tienda 4: ${ingreso_total_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
# Calcular las ventas por categoría para cada tienda
ventas_por_categoria_tienda1 = tienda1_df['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda2 = tienda2_df['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda3 = tienda3_df['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda4 = tienda4_df['Categoría del Producto'].value_counts()
```

```python

# Mostrar los resultados para cada tienda
print("Ventas por categoría de la Tienda 1:\n", ventas_por_categoria_tienda1)
print("\nVentas por categoría de la Tienda 2:\n", ventas_por_categoria_tienda2)
print("\nVentas por categoría de la Tienda 3:\n", ventas_por_categoria_tienda3)
print("\nVentas por categoría de la Tienda 4:\n", ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la valoración media para cada tienda
valoracion_media_tienda1 = tienda1_df['Calificación'].mean()
valoracion_media_tienda2 = tienda2_df['Calificación'].mean()
valoracion_media_tienda3 = tienda3_df['Calificación'].mean()
valoracion_media_tienda4 = tienda4_df['Calificación'].mean()

```

```python
# Mostrar los resultados
print(f"Valoración media de la Tienda 1: {valoracion_media_tienda1:.2f}")
print(f"Valoración media de la Tienda 2: {valoracion_media_tienda2:.2f}")
print(f"Valoración media de la Tienda 3: {valoracion_media_tienda3:.2f}")
print(f"Valoración media de la Tienda 4: {valoracion_media_tienda4:.2f}")
```

# 5. Envío promedio por tienda

```python
# Calcular el valor del envío promedio para cada tienda
envio_promedio_tienda1 = tienda1_df['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2_df['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3_df['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4_df['Costo de envío'].mean()

```

# 11. ¡Extra! Análisis del desempeño geográfico

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar valores nulos en latitud y longitud para evitar errores
tienda1_df = tienda1_df.dropna(subset=["lat", "lon"])
tienda2_df = tienda2_df.dropna(subset=["lat", "lon"])
tienda3_df = tienda3_df.dropna(subset=["lat", "lon"])
tienda4_df = tienda4_df.dropna(subset=["lat", "lon"])

# Gráfico de dispersión con coordenadas corregidas
plt.figure(figsize=(10,6))
sns.scatterplot(x=tienda1_df["lon"], y=tienda1_df["lat"], alpha=0.6, color="blue", label="Tienda 1")
sns.scatterplot(x=tienda2_df["lon"], y=tienda2_df["lat"], alpha=0.6, color="red", label="Tienda 2")
sns.scatterplot(x=tienda3_df["lon"], y=tienda3_df["lat"], alpha=0.6, color="green", label="Tienda 3")
sns.scatterplot(x=tienda4_df["lon"], y=tienda4_df["lat"], alpha=0.6, color="purple", label="Tienda 4")

# Ajustar etiquetas y título
plt.xlabel("Longitud", fontsize=12, fontweight='bold')
plt.ylabel("Latitud", fontsize=12, fontweight='bold')
plt.title("Distribución Geográfica de Ventas por Tienda", fontsize=14, fontweight='bold')
plt.legend()
plt.show()
```

```python
import folium
from folium.plugins import HeatMap

# Crear mapa interactivo centrado en la ubicación promedio de las tiendas
centro_lat = pd.concat([tienda1_df["lat"], tienda2_df["lat"], tienda3_df["lat"], tienda4_df["lat"]]).mean()
centro_lon = pd.concat([tienda1_df["lon"], tienda2_df["lon"], tienda3_df["lon"], tienda4_df["lon"]]).mean()
m = folium.Map(location=[centro_lat, centro_lon], zoom_start=10)

# Lista de coordenadas (latitud, longitud) de todas las ventas
ventas_coordenadas = pd.concat([
    tienda1_df[["lat", "lon"]],
    tienda2_df[["lat", "lon"]],
    tienda3_df[["lat", "lon"]],
    tienda4_df[["lat", "lon"]]
]).dropna().values.tolist()

# Agregar mapa de calor al mapa interactivo
HeatMap(ventas_coordenadas, radius=10).add_to(m)

# Mostrar mapa interactivo
m
```
